```python
import pandas as pd
```

```python
df=pd.read_excel("Car data saved-1.xlsx")
df
```

```python
pivot=df.pivot_table(index='Name',columns='Type',values='Onroad Price',aggfunc='min')
pivot
```

```python
pivot1=df.pivot_table(
    index='Name',columns='Type',values=['Onroad Price','Ex-Showroom Price'],aggfunc='min').fillna(0)
pivot1
```

```python
preference=pd.DataFrame(
    {
        'preference':['Yes','No'],
                      'Type':['Petrol','Diesel']
    })
preference
```

```python
merging=pd.merge(df,preference,on='Type',how='right')
merging
                 
```
