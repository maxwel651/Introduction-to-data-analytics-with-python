```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('Car.xlsx')
df
```

```python
df.columns
```

```python
variables=['Min Price (Lakh)', 'Max Price (Lakh)', 'Range (kmpl)', 'CC',
       'Seats', 'Variants','Ex-Showroom Price', 'RTO', 'Insurance',
       'Other', 'Onroad Price']

```

```python
plt.figure(figsize=(15, 12))
for i, var in enumerate(variables, start=1):
    plt.subplot(4, 3, i)
    sns.scatterplot(x=df[var],y=df['Max Price (Lakh)'],alpha=0.7)
    plt.title(f'Scatter plot of {var} vs onroad price')
    plt.xlabel(var)
    plt.ylabel('Target')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 10))
for i, var in enumerate(variables, start=1):
    plt.subplot(3, 4, i)
    sns.histplot(df[var], bins=20, kde= True)
    plt.title(f'Histogram of {var}')
    plt.xlabel(var)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(data=df,x='Type',y='Range (kmpl)',color='yellow')
plt.title('Boxplot of Range by Car Type')
plt.show()
```

```python
selected_columns=df[['Min Price (Lakh)', 'Max Price (Lakh)', 'Range (kmpl)', 'CC',
       'Seats', 'Variants','Ex-Showroom Price', 'RTO', 'Insurance',
       'Other']]
selected_columns.head()
```

```python
corr=selected_columns.corr()
corr
```

```python
#plotting heatmap
```

```python
sns.heatmap(corr,annot=True,cmap='Reds')
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python
#violin plot of max price by car type
```

```python
sns.violinplot(data=df,x="Type",y="Max Price (Lakh)")
plt.title('Violin Plot of Maximum Price by Car Type')
plt.show()
```
